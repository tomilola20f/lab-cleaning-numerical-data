```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
df
```

```python
df = df.drop(columns=["Customer"], axis=1)
df
```

```python
df_num=df.select_dtypes(include=np.number)
df_cat=df.select_dtypes(include=np.object)
```

```python
df_cat
```

```python
for cat in df_cat:
    sns.boxplot(x = df_cat[cat], y =df['Customer Lifetime Value'])
    plt.show()
```

```python
sns.boxenplot(data=df, x=df_cat[cat], y=df['Customer Lifetime Value'])
```
